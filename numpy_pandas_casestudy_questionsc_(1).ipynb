2. Extract all the **even numbers** from the array.  
3. Replace all numbers **greater than 50** with **0**.  
4. Extract the diagonal elements of the array.

```python
# Solution for Question 2
import numpy as np
array = np.random.randint(10,100,size=(5,5))
print(array)
# Your code here
evennumber = array[array % 2==0]
print("Even Number:")
print(evennumber)
array[array>50]=0
print("Array after replacing")
print(array)
daigonal_elements = np.diag(array)
print("Daigonal elements:")
print(daigonal_elements)
```

## ðŸ§  **Question 3: DataFrame Creation and Operations (Pandas)**
1. Create a Pandas DataFrame containing the following data:

| Student_ID | Name   | Subject | Marks |
|------------|--------|---------|-------|
| 1          | Alice  | Math    | 88    |
| 2          | Bob    | Science | 75    |
| 3          | Carol  | History | 92    |
| 4          | David  | Math    | 85    |
| 5          | Eva    | Science | 90    |

2. Find the **average marks** for each subject.  
3. Add a new column named **Grade** based on the marks:
- Marks â‰¥ 90 â†’ A
- Marks â‰¥ 80 â†’ B
- Marks â‰¥ 70 â†’ C
- Else â†’ D

```python
# Solution for Question 3
import pandas as pd
data={"student_id":[1,2,3,4,5],
      "name":["alice","bob","carol","david","eva"],
      "subject":["math","science","history","math","science"],
      "mark":[88,75,92,85,90]}
df=pd.DataFrame(data)
print("data:\n",df)


# Your code here
```

```python
average_marks=df.groupby("subject")["mark"].mean()
print(average_marks)
```

```python
def assign_grade(mark):
  if mark>=90:
    return"A"
  elif mark>=80:
    return"B"
  elif mark>=70:
    return"C"
  else:
    return"D"
df["grade"]=df["mark"].apply(assign_grade)
print("\nDataFrame with grade:\n",df)
```

## ðŸ§  **Question 4: Dataset Analysis (Pandas, Matplotlib, Seaborn)**
Use the dataset `diabetes.csv` and perform the following tasks:

1. Display the **BMI** and **Blood Pressure** for samples with **index ranging from 20 to 40**.  
2. Find how many patients have **BMI > 30**.  
3. Find the **mean and standard deviation** of the **Glucose** levels.  
4. Plot a **boxplot** for the column **Insulin**.  
5. Create a **heatmap** showing the correlation between the numerical columns in the dataset.

```python
# Solution for Question 4
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('/content/diabetes_model (1).csv')
print(data)
```

```python
print("BMI and Blood Presure for index 20 to 40:\n",data.loc[20:40],["BMI"],["Blood pressyre"])
```

```python
glucose_mean=data["Glucose"].mean()
glucose_std=data["Glucose"].std()
print("\nMean of Glucose levels:",glucose_mean)
print("Standard deviation of glucose levels:",glucose_std)
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(data["Insulin"])
plt.title("Boxplot of Insulin")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(data.corr(),annot=True,cmap="coolwarm",fmt="2f")
plt.title("coorelation heatmap of numerical columns")
plt.show()
```
